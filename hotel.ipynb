Приводим названия колонок к нижнему регистру и заменяем пробелы на знак нижнего подчеркивания.

```python
def replace_and_low(name):
    name = name.replace(' ', '_').lower()
    return name

df = df.rename(columns = replace_and_low)
df
```

На сколько ночей в среднем бронируют отели типа City Hotel? Resort Hotel? Запишем полученные значения в пропуски с точностью до 2 знаков после точки.

```python
nights_in_hotel = df \
    .query('hotel == "City Hotel" or hotel == "Resort Hotel"') \
    .groupby('hotel', as_index = False).agg({'stays_total_nights' : 'mean'}) \
    .round(2)
nights_in_hotel
```

Необходимо посчитать, сколько клиентов было потеряно в процессе. Иными словами, посчитать метрику под названием Churn Rate.

В нашем случае Churn Rate - это процент клиентов, которые отменили бронирование. Давайте посмотрим, как эта метрика связана с наличием детей у клиентов!

Создадим переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Далее проверим, среди какой группы пользователей показатель оттока выше. 

```python
df['has_kids'] = df.total_kids != 0
churn_rate_with_kids = df.query('has_kids == True and is_canceled == 1').shape[0] *100 / df.query('has_kids == True').shape[0]
churn_rate_with_kids = round(churn_rate_with_kids,2)
churn_rate_with_kids
```

```python
churn_rate_without_kids = df.query('has_kids == False and is_canceled == 1').shape[0] *100/ df.query('has_kids == False').shape[0]
churn_rate_without_kids = round(churn_rate_without_kids,2)
churn_rate_without_kids
```
